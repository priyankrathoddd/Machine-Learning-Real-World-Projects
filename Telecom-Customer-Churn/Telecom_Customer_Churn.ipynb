```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
# Load DataSet

df = pd.read_csv("Telco-Customer-Churn.csv")
df.head()
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
# Basic Data Cleaning

df = df.drop(['customerID'],axis=1)
```

```python
# Check Delete Columns

df.columns
```

```python
# Fix TotalCharges Column DataType To Numeric

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
```

```python
print(df.isnull().sum())
```

```python
df["Churn"] = df["Churn"].map({"No":0,"Yes":1})
df.head()
#print(df.isnull().sum())
```

```python
# Feature and Target

X = df.drop(['Churn'],axis=1)
y = df['Churn']
```

```python
X.head(2)
```

```python
y.head()
```

```python
df.columns
```

```python
categorical_cols = ['gender','SeniorCitizen','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity',
'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','StreamingMovies', 'Contract', 'PaperlessBilling','PaymentMethod']

numerical_cols = ['tenure','MonthlyCharges','TotalCharges']

# Preprocessing

preprocessor = ColumnTransformer(transformers=[('num',StandardScaler(),numerical_cols),
    ('cat',OneHotEncoder(drop='first',sparse_output=False),categorical_cols)])
```

```python
# Train , Test , Split
# Test 20%
# Train 80%

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)

print('Train data X',X_train.shape)
print('Train data y',y_train.shape)

print('Test data X',X_train.shape)
print('Test data y',y_train.shape)
```

```python
# Create a model pipeline

model = Pipeline(steps=[('preprocessor',preprocessor),('classifier',LogisticRegression(max_iter=1000,random_state=42))])

# Train the model

model.fit(X_train,y_train)
```

```python
# Make Prediction

y_pred = model.predict(X_test)
print("Accuracy : ",accuracy_score(y_test,y_pred))
print("Classification report : \n",classification_report(y_test,y_pred))
print("Confusion matrix \n",confusion_matrix(y_test,y_pred))
```
